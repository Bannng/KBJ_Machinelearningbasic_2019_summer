Initializing x_bar and weight v and w

```
#initialize part

one_row = np.ones([len(x1),1])
x1_transpose = x1.reshape(-1,1)
x2_transpose = x2.reshape(-1,1)

#making x_bar
x_bar_1 = np.hstack([one_row,x1_transpose])
x_bar = np.hstack([x_bar_1,x2_transpose])

#setting weight
#weight_v = np.empty([3,5],dtype=float)
#weight_w = np.empty([6,2],dtype=float)
weight_v = np.random.rand(3,5)
weight_w = np.random.rand(6,2)

print(weight_v)
print(weight_w)
```

Define functions

```
def activationfun(input_arr):
  input_exp = np.exp(-1*input_arr)
  input_exp_plusone = 1+input_exp
  input_exp_plusone_div = 1/input_exp_plusone
  return input_exp_plusone_div

def makebarmat(input_arr):
  one_row = np.ones([len(input_arr),1])
  barmat = np.hstack([one_row,input_arr])
  return barmat

def geterr(g, real_y):
  value1 = 0
  for i in range(0,len(real_y)):
    for j in range(0,len(real_y[0])):
      value1 += (g[i,j]-real_y[i,j])**2
  err = 0.5*value1    
  return err

def dobackpropagation_for_w(weight_w, alpha, g, real_y, f_bar):  
  for k in range(0,len(weight_w)):
    for j in range(0,len(real_y[0])):
      value_sum = 0
      for i in range(0,len(real_y)):
        value_sum += (g[i,j]-real_y[i,j])*g[i,j]*(1-g[i,j])*f_bar[i,k]
      weight_w[k,j] = weight_w[k,j] - alpha*value_sum
  return weight_w

def dobackpropagation_for_v(weight_v, alpha, g, real_y, w, f, x_bar):  
  for n in range(0,len(weight_v)):
    for k in range(0,len(weight_v[0])):
      value_sum = 0
      for i in range(0,len(real_y)):
        for j in range(0,len(real_y[0])):
          value_sum += (g[i,j]-real_y[i,j])*g[i,j]*(1-g[i,j])*w[k,j]*f[i,k]*(1-f[i,k])*x_bar[i,n]
      weight_v[n,k] = weight_v[n,k] - alpha*value_sum
  return weight_v   

```

Main loop (forward porpagaton and back propagation)

```
E_present = 0.5
E_previous = 1
e = 0.000001
new_weight_w = weight_w
new_weight_v = weight_v
learningrate = 0.01
count = 0
err_arr = np.zeros(1,dtype=float)


#for epoch in range(0,10000):
while(np.abs(E_present-E_previous)>e):
  # forward propagation 
  # x_bar to x_bar_bar
  x_bar_bar = np.matmul(x_bar,new_weight_v)
  #pass through activation
  f = activationfun(x_bar_bar)
  # second hidden layer
  f_bar = makebarmat(f)
  #f_bar to f_bar_bar
  f_bar_bar = np.matmul(f_bar,new_weight_w)
  g = activationfun(f_bar_bar)
  #updating Errors
  E_previous = E_present
  E_present = geterr(g, y_real)
  #back propagation
  new_weight_w = dobackpropagation_for_w(weight_w, learningrate, g, y_real, f_bar)
  new_weight_v = dobackpropagation_for_v(weight_v, learningrate, g, y_real, weight_w, f, x_bar)
  #making err array to plot
  err_arr = np.append(err_arr , E_present)
  count += 1
  

  
# checking result  
g



    
  
```

Get Error Plot

```
count_arr = np.arange(0,count)
real_err_arr = err_arr[1:]
plt.xlabel('iteration')
plt.ylabel('Error')
plt.plot(count_arr , real_err_arr , '-')
count_arr.shape , real_err_arr.shape
```
